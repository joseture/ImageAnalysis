```python
from scipy.ndimage import correlate
from skimage import color, io
import matplotlib.pyplot as plt
import numpy as np

## Exercise 1 & 2

input_image = np.arange(25).reshape(5,5)
print(input_image)

weights = [[0, 1, 0],
		   [1, 2, 1],
		   [0, 1, 0]]

# res_img = correlate(input_image, weights)

res_img = correlate(input_image, weights, mode="constant", cval=10)

print(input_image[3, 3])
print(res_img[3, 3])
```

```python
## Exercise 3 (Mean filter)

path = "data/"

img = io.imread(path + "Gaussian.png")

size = 20
# Two dimensional filter filled with 1
weights = np.ones([size, size])
# Normalize weights
weights = weights / np.sum(weights)

res_img = correlate(img, weights, mode="constant", cval=10)
 
io.imshow(res_img)
io.show()
```

```python
## Exercise 4 (Median filter)
from skimage.filters import median

size = 20
footprint = np.ones([size, size])
med_img = median(img, footprint)

io.imshow(med_img)
io.show()
```

```python
## Exercise 5
path = "data/"
img = io.imread(path + "SaltPepper.png")

# Mean filter
size = 5
weights = np.ones([size, size])
weights = weights / np.sum(weights)

res_img = correlate(img, weights, mode="constant", cval=10)

# Median filter
size = 5
footprint = np.ones([size, size])
med_img = median(img, footprint)

io.imshow(res_img)
io.show()
```

```python
## Exercise 10
from skimage.filters import threshold_otsu


path = "data/"
img = io.imread(path + "ElbowCTSlice.png")

# gaussian filter
sigma = 2
gauss_img = gaussian(img, sigma)

# median filter

size = 10
footprint = np.ones([size, size])
med_img = median(img, footprint)

# gradients

prew_h = prewitt_h(gauss_img)
prew_v = prewitt_v(gauss_img)
prew= prewitt(med_img)

# otsu's thresholding

thres_otsu = threshold_otsu(prew)

# Binary image

thres_bin = prew < thres_otsu


io.imshow(thres_bin)
io.show()

```
